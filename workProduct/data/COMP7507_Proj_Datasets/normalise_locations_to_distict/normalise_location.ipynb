```python
import pandas as pd

# District coordinates
district_coords = {
    "Central and Western": (22.2833, 114.1500),
    "Eastern": (22.2838, 114.2250),
    "Southern": (22.2475, 114.1588),
    "Wan Chai": (22.2760, 114.1800),
    "Kowloon City": (22.3282, 114.1913),
    "Kwun Tong": (22.3136, 114.2250),
    "Sham Shui Po": (22.3300, 114.1600),
    "Wong Tai Sin": (22.3400, 114.2000),
    "Yau Tsim Mong": (22.3200, 114.1700),
    "Islands": (22.2700, 113.9500),
    "Kwai Tsing": (22.3600, 114.1000),
    "North": (22.5000, 114.1300),
    "Sai Kung": (22.3800, 114.2700),
    "Sha Tin": (22.4000, 114.2000),
    "Tai Po": (22.4500, 114.1700),
    "Tsuen Wan": (22.3700, 114.1100),
    "Tuen Mun": (22.3900, 113.9800),
    "Yuen Long": (22.4500, 114.0300)
}

# Mapping station/location names to districts
# station_to_district = {
#     # Sha Tin
#     "SHATIN": "Sha Tin", "Sha Tin": "Sha Tin", "SHA": "Sha Tin", "SKG": "Sha Tin",

#     # Tsuen Wan
#     "TSUEN WAN": "Tsuen Wan", "Tai Mo Shan": "Tsuen Wan", "SE1": "Tsuen Wan", "SSP": "Tsuen Wan", "TW": "Tsuen Wan", "TWN": "Tsuen Wan",

#     # Central and Western
#     "CENTRAL": "Central and Western", "CENTRAL/WESTERN": "Central and Western", "The Peak": "Central and Western",
#     "Central Pier": "Central and Western", "Green Island": "Central and Western",

#     # Eastern
#     "EASTERN": "Eastern", "Shau Kei Wan": "Eastern", "NGP": "Eastern", "SKW": "Eastern", "North Point": "Eastern", "Shell Oil Depot": "Eastern",

#     # Kwun Tong
#     "KWUN TONG": "Kwun Tong",

#     # Tuen Mun
#     "TUEN MUN": "Tuen Mun", "Tuen Mun Children and Juvenile Home": "Tuen Mun", "TMS": "Tuen Mun", "Tuen Mun Government Offices": "Tuen Mun",

#     # Islands
#     "TUNG CHUNG": "Islands", "Cheung Chau": "Islands", "Peng Chau": "Islands", "Waglan Island": "Islands", "Lamma Island": "Islands",
#     "Sha Chau": "Islands", "Bluff Head": "Islands", "Cheung Chau Beach": "Islands", "KP": "Islands", "KLT": "Islands",
#     "TY1": "Islands", "TYW": "Islands", "VP1": "Islands", "WGL": "Islands", "CCH": "Islands", "HKA": "Islands", "TPO": "Islands",

#     # Sham Shui Po
#     "SHAM SHUI PO": "Sham Shui Po",  "Sham Shui Po" : "Sham Shui Po", "SSP" : "Sham Shui Po",

#     # Southern
#     "SOUTHERN": "Southern", "HPV": "Southern", "Wong Chuk Hang": "Southern",

#     # Yuen Long
#     "YUEN LONG": "Yuen Long", "Lau Fau Shan": "Yuen Long", "Shek Kong": "Yuen Long", "JKB": "Yuen Long", "KSC": "Yuen Long",
#     "LFS": "Yuen Long", "TU1": "Yuen Long", "WLP": "Yuen Long", "YCT": "Yuen Long", "YLP": "Yuen Long", "Wetland Park": "Yuen Long",

#     # North
#     "NORTH": "North", "Sheung Shui": "North", "Ta Kwu Ling": "North", "SEK": "North", "STY": "North", "TKL": "North",

#     # Kwai Tsing
#     "KWAI CHUNG": "Kwai Tsing", "Ching Pak House(Tsing Yi)": "Kwai Tsing",

#     # Sai Kung
#     "TSEUNG KWAN O": "Sai Kung", "Tseung Kwan O": "Sai Kung", "Kau Sai Chau": "Sai Kung", "Pak Tam Chung (Tsak Yue Wu)": "Sai Kung",
#     "HKP": "Sai Kung", "SSH": "Sai Kung",

#     # Tai Po
#     "TAI PO": "Tai Po", "Tai Mei Tuk": "Tai Po", "Tap Mun East": "Tai Po", "Tai Po Kau": "Tai Po", "TC": "Tai Po",
#     "TAP MUN" : "Tai Po", "Tap Mun" : "Tai Po",

#     # Yau Tsim Mong
#     "MONG KOK": "Yau Tsim Mong", "King's Park": "Yau Tsim Mong", "HKO": "Yau Tsim Mong", "HKS": "Yau Tsim Mong",
#     "Star Ferry(Kowloon)": "Yau Tsim Mong", "(mm) at the Hong Kong Observatory": "Yau Tsim Mong",

#     # Wan Chai
#     "CAUSEWAY BAY": "Wan Chai", "CWB": "Wan Chai",

#     # Kowloon City
#     "Kai Tak": "Kowloon City", "KTG": "Kowloon City",

#     # Wong Tai Sin
#     "Tate's Cairn": "Wong Tai Sin", "WTS": "Wong Tai Sin"
# }

import json
station_to_district = json.load(open("./station_to_district.json", mode='r'))

# Function to normalize and add coordinates
def normalize_district(df, column):
    df['norm_dist'] = df[column].map(station_to_district)
    df['dist_lat'] = df['norm_dist'].map(lambda x: district_coords.get(x, (None, None))[0])
    df['dist_long'] = df['norm_dist'].map(lambda x: district_coords.get(x, (None, None))[1])
    return df

# Apply normalization
particles_df = normalize_district(particles_df, 'location')
rainfall_df = normalize_district(rainfall_df, 'locationName')
temperature_df = normalize_district(temperature_df, 'station')
windspeed_df = normalize_district(windspeed_df, 'locationName')
```

## Data Cleaning for Specific Columns

### Particles

```python
particles_df['value'] = particles_df['value'].replace("N.A.", "")
particles_df['date'] = pd.to_datetime(particles_df[['year', 'month', 'day']])

# Convert 'value' column to numeric, treating empty strings as NaN
particles_df['value'] = pd.to_numeric(particles_df['value'], errors='coerce')

particles_df['value_abs'] = particles_df.apply(
    lambda row: row['value'] * 10 if (row['pollutantType'] == 'Carbon Monoxide' and pd.notnull(row['value'])) else row['value'], axis=1
)

particles_df.to_csv('./cleaned/particles_norm_dist_clean.csv')
```

```python
rainfall_df['value'] = rainfall_df['value'].replace("N.A.", "")
rainfall_df['value'] = pd.to_numeric(rainfall_df['value'], errors='coerce')
rainfall_df['date'] = pd.to_datetime(rainfall_df[['year', 'month', 'day']], errors='coerce')
rainfall_df.to_csv('./cleaned/rainfall_norm_dist_clean.csv')
```

```python
#temperature_df['value'] = temperature_df['value'].replace("N.A.", "")
temperature_df['date'] = pd.to_datetime(temperature_df[['year', 'month', 'day']], errors='coerce')
temperature_df['minTemp'] = pd.to_numeric(temperature_df['minTemp'], errors='coerce')
temperature_df['meanTemp'] = pd.to_numeric(temperature_df['meanTemp'], errors='coerce')
temperature_df['maxTemp'] = pd.to_numeric(temperature_df['maxTemp'], errors='coerce')
temperature_df.to_csv('./cleaned/temperature_norm_dist_clean.csv')
```

```python
windspeed_df['value'] = windspeed_df['value'].replace("N.A.", "")
windspeed_df['date'] = pd.to_datetime(windspeed_df[['year', 'month', 'day']], errors='coerce')
windspeed_df['value'] = pd.to_numeric(windspeed_df['value'], errors='coerce')
windspeed_df.to_csv('./cleaned/windspeed_norm_dist_clean.csv')
```

### Separate Pollutant Sheets

```python
pollutants = particles_df['pollutantAcronym'].unique()
print(pollutants)
```

```python
dataframes_dict: dict[str, pd.DataFrame] = dict()
for pollutant in pollutants:
    dataframes_dict[f"pollutant{pollutant}_df"] = particles_df[particles_df['pollutantAcronym'] == pollutant]

dataframes_dict["particles_df"] = particles_df
dataframes_dict["rainfall_df"] = rainfall_df
dataframes_dict["temperature_df"] = temperature_df
dataframes_dict["windspeed_df"] = windspeed_df
```

```python
import pandas as pd
import os

xlsx_file_name = 'master_data_v1.xlsx'
xlsx_file_path = os.path.join(os.getcwd(), 'xlsx', xlsx_file_name)

# Create a Pandas Excel writer using XlsxWriter as the engine
with pd.ExcelWriter(f'{xlsx_file_path}', engine='xlsxwriter') as writer:
    # Write each dataframe to a different worksheet
    for key in dataframes_dict:
        dataframe = dataframes_dict[key]
        dataframe.to_excel(writer, sheet_name=f'{key}', index=False)

print(f"DataFrames have been saved to {xlsx_file_name} with each DataFrame in a separate sheet.")
```
